# **5.Type casting in numpy array**

```python
#float to int :

arr3=np.array([1.2,2.3,3.5,4])
print(arr3)
#convert into int
arr4= arr3.astype('i')
print(arr4)
```

```python
#int to float

arr5=np.array([1,2,3,3,4])
print(arr5)
#convert into int
newarr5= arr4.astype('f')
print(newarr5)
```

```python
a=arr4.astype(float) # 'f or type only float

print(a)
```

```python
#int to boolean

arr_int = np.array([0, 1, 2, 3, 0])

# Type casting to boolean
arr_bool = arr_int.astype(bool)

print(arr_bool)
```

```python
# int to string :
arr_int = np.array([1, 2, 3, 4])

# Type casting to string
arr_str = arr_int.astype(str)

print(arr_str)
```

**full()**
- return same elements

```python
s=np.full((2,2),-3) # 2d 2rows 2 columns of 4 elements
print(s)
print(s.ndim)
```

```python
s=np.full((3,2),4) # 2d 3rows 2 columns of 4 all elements
print(s)
print(s.ndim)
```

```python
s=np.full((2),4) # 1d
print(s)
print(s.ndim)
```

```python
s=np.full((3),5) # 1d 5 occurs 3 times
print(s)
print(s.ndim)
```

```python
s=np.full(5,(3)) # 1d ,3 occurs 5 times
print(s)
print(s.ndim)
```

# **7.Reshape of an array**
- change dimensions

```python
ar=np.array([1,2,3,4,5,6])
print(ar)
print(ar.ndim)
```

```python
ar1=ar.reshape(3,2) # 1d to 2d
print(ar1)
print(ar1.ndim)
```

```python
# 1d to 3d
ar2=np.array([1,2,3,5,6,7])
print(ar2)
print(ar2.ndim)
ar3=ar2.reshape(2,3,1)
print(ar3)
print(ar3.ndim)
```

```python
a2=a.reshape(2,2,3)
print(a2)
print(a2.ndim)
```

```python
# create 1 to 17 elements of 3d array with 2 layers

d=np.arange(1,17)
print(d)
print(d.ndim)
```
