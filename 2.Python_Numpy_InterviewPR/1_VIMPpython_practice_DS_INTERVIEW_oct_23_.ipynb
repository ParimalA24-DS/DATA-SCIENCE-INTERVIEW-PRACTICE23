## 3.Palindrome ? or not ?


*This program checks if the input is the same when read forwards and backwards, ignoring spaces, commas, and periods*

```python
def is_palindrome(word):
    reversed_word = word[::-1]
    return word == reversed_word

# Get user input
word = input("Enter a word: ")

if is_palindrome(word):
    print(f"'{word}' is a palindrome")
else:
    print(f"'{word}' is not a palindrome")
```

### Use pop():
When you want to remove and retrieve an arbitrary element from the set. This is often used when you want to process elements in a random order.
### Use remove():
When you want to remove a specific element and you're certain it exists in the set. If the element is not present, it raises an error.
### Use discard():
When you want to remove a specific element, but it's acceptable for the element not to be present in the set. **If the element is not present, no error is raised**.
