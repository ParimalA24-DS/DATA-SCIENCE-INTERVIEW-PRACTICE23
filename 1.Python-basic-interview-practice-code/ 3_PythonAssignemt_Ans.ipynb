**list comrehension and Generator differences?**

```python
## Using a list comprehension
squares_list = [i**2 for i in range(1, 11)]
print("List of squares:", squares_list)

#The list comprehension [i**2 for i in range(1, 11)] creates a list with all the square numbers from 1 to 10.
```

```python
# Using a generator expression
squares_generator = (i**2 for i in range(1, 11))
print("Generator of squares:", squares_generator)

# To get the values, we can iterate over the generator
for square in squares_generator:
    print(square,end=" ")
```

*The key difference is that the list comprehension immediately computes and stores all the square numbers in memory, while the generator expression doesn't compute any squares until you actually iterate over it.*
