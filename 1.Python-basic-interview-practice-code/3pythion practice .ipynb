#list: combine two list and give odd list numbers from list

```python
a=[12,34,343, 545]

b=[34,53,35,64]


def odd_list(l1,l2):
  c=[]
  for i in l1:
    if i%2!=0:
      c.append(i)
  for j in l2:
    if j%2!=0:
      c.append(j)

  return(c)

odd_list(a,b)
```

**Hashability: Tuples are hashable, which means they can be used as keys in dictionaries or as elements in sets. The immutability guarantees that the hash value of a tuple remains constant. explain with easy simple example**

```python

student_grades = {}

# Add entries to the dictionary
student_grades[('John', 'Doe')] = 85
student_grades[('Jane', 'Doe')] = 92
student_grades[('Bob', 'Smith')] = 78
```

```python
student_grades
```

***heree'using tuples as keys. Each key is a tuple containing the first name and last name of a student. This works because tuples are immutable, and their contents won't change. Therefore, the hash value of each tuple remains constant.***

```python
# Retrieve Jane Doe's grade
jane_grade = student_grades[('Jane', 'Doe')]
print(jane_grade)
```

**Find the Factorial of a Number:**

```python
n = int(input("Enter a number: "))

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)


result = factorial(n)
print(f"The factorial of {n} is: {result}")
```

**string :palindrome or not **

```python
user_input = input("Enter a string: ")
def is_palindrome(input_str):
    return input_str == input_str[::-1]


if is_palindrome(user_input):
    print("Yes, it's a palindrome!")
else:
    print("No, it's not a palindrome.")
```

**Write a program to find the sum of squares of first n natural
numbers**



```python
n = int(input("Enter a positive integer (n): "))

def sum_of_squares(n):
    sum = 0
    for i in range(1, n+1):
        sum =sum+(i**2)
    return sum


result = sum_of_squares(n)

print(f"The sum of squares of the first {n} natural numbers is: {result}")
```
