```python
# list check even and odd :
l=[1,2,3,4,5]
def check_oe(l):
  for i in l:
    if i%2==0:
      print("number is even",i)
    else:
      print("odd no.",i)

check_oe(l)
```

```python
#if no. is 450 skip that number , n>700 exit loop

l1=[100,200,300,250,450,122,133,450,345,567,800,900,945]

for i in l1:
  if i==450:
    continue
  elif i>700:
    break
  elif i%5==0:
    print(f"this number{i}is divisible by 5")
```
