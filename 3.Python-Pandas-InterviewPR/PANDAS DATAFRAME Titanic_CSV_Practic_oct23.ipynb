# Grouping the data:

*Grouping data in Pandas is a powerful operation that allows you to split data into groups based on some criteria, apply a function to each group independently, and then combine the results back into a DataFrame*

*When to Use groupby:
Aggregation: When you want to compute summary statistics (like mean, median, count, etc.) for different groups in your dataset.

Comparative Analysis: When you want to compare the characteristics or statistics of different groups.

Data Exploration: To gain insights and explore patterns within specific subsets of the data.

Conditional Operations: When you want to apply different operations to different groups based on specific conditions*

```python
#Group Pclass column
d2.groupby(data['Pclass']).count()

#this count() method counts the number of occurrences of each unique value after grouping.
# how many occurrences are there for each unique value in the 'Pclass' column?"
```

```python
#"What is the maximum fare paid by passengers in each class?
d2.groupby(data['Pclass']).max('Fare')
# Pcass1 :512fare charges
```

```python
d4=d2.groupby(data['Age']).describe()
d4.head()
#What are the summary statistics for different age groups?"
```

```python
# What is the average value of each numeric variable for different gender groups?"
d= d2.groupby(data['Sex']).mean()
d.head()
```

```python
d3=d2.groupby(data['Pclass']).agg({'Fare':'max'})
d3.head()
#That is the maximum fare paid by passengers in each class?"
# The output shows the maximum fare for each class.
```

*ThIS code groups the data by passenger class and then calculates the highest fare within each class**

#Joining the dataframes-

```python
!. # pd.concat():
## This method concatenates DataFrames along a particular axis (either rows or columns).
# It's useful when you want to stack DataFrames on top of each other or side by side.
import pandas as pd

df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                    'B': ['B0', 'B1', 'B2'],
                    'C': ['C0', 'C1', 'C2']})

df2 = pd.DataFrame({'A': ['A3', 'A4', 'A5'],
                    'B': ['B3', 'B4', 'B5'],
                    'C': ['C3', 'C4', 'C5']})

```

```python
df1
```

```python
df2
```

```python
result = pd.concat([df1, df2], axis=0) # Concatenating along rows rows are added
result
```

```python
result1 = pd.concat([df1, df2], axis=1) # This method concatenates DataFrames side by side
result1 # columnadded
```

**II.pd.merge():
This method performs SQL-style joins on DataFrames based on one or more keys. It's useful when you want to combine DataFrames based on a common column**

```python
df1['key_column'] = 1
df2['key_column'] = 1
```

```python
df1
```

```python
result2 = pd.merge(df1, df2, on='key_column', how='inner')
```

**III. df.join():
This method is used for combining the columns of two potentially differently-indexed DataFrames into a single DataFrame. It's useful when you want to join based on the index.**

```python
 #Removing the 'key_column' after the merge
result2.drop(columns=['key_column'], inplace=True)
```

*you are joining on indices and both DataFrames have the same index, join() is more concise and convenient.*
*If you need to join on columns with different names or you want more flexibility in specifying the join columns, pd.merge() is more appropriate.* In general, pd.merge() is more versatile and commonly used for a wider range of merge operations.

```python
result3 = df1.join(df2, how='inner')
result3
```
