##**Q.2 Merging two datframes differant differant ways**

```python
#crete two dictionaries first

di1 = {'Name': ['Alice', 'Bob', 'Charlie'],
         'Score1': [85, 90, 75]}

di2 = {'Name': ['David', 'Eve', 'Frank'],
         'Score2': [70, 80, 95]}
```

```python
# Creating DataFrames from the dictionaries
df1 = pd.DataFrame(di1)
df2 = pd.DataFrame(di2)
```

```python
df1
```

```python
df2
```

```python
# Merging DataFrames based on 'Name'
merged_df = pd.merge(df1, df2, on='Name', how='inner')

print(merged_df) # no common element so choose another
```

```python
#differet dicts:
di1 = {'Name': ['Alice', 'Bob', 'Charlie'],
       'Score1': [85, 90, 75]}

di2 = {'Name': ['Charlie', 'Eve', 'Frank'],
       'Score2': [70, 80, 95]}

df1 = pd.DataFrame(di1)
df2 = pd.DataFrame(di2)
```

```python
r=pd.merge(df1, df2, on='Name', how='inner')
```

```python
r # coomon column charlie so only shows his info.
```

```python
# Left join merge :
# Left Join
left_merged = pd.merge(df1, df2, on='Name', how='left')
print(left_merged)
```

```python
#A left join returns all the rows from the left DataFrame (df1), and the matched rows from the right DataFrame (df2).
#If there are no matching rows in df2, it will fill the values with NaN.
```

```python
#Right Join
right_merged = pd.merge(df1, df2, on='Name', how='right')
right_merged
```

```python
#A right join returns all the rows from the right DataFrame (df2), and the matched rows from the left DataFrame (df1).
#If there are no matching rows in df1, it will fill the values with NaN.
```

```python
# Outer Join:
outer_merged = pd.merge(df1, df2, on='Name', how='outer')
outer_merged
```

```python
# Full Outer Join:or outer
#An outer join returns all the rows when there is a match in one of the DataFrames.
#It fills non-matching rows with NaN values.
```
