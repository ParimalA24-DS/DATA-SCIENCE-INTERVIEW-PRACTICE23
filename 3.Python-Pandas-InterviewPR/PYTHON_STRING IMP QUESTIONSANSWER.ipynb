**Q5.Write a Python program to capitalize the first letter of each word in a sentence**

```python
s = "asd hello how are you"
capitalized_sentence = ' '.join(word.capitalize() for word in s.split())
print(capitalized_sentence)
```

**Q6.Write a Python program to check if two strings are anagrams.**

```python
def are_anagrams(str1, str2):
    # Remove spaces and convert both strings to lowercase
    str1 = str1.replace(' ', '').lower()
    str2 = str2.replace(' ', '').lower()

    # Check if the sorted characters of both strings are equal
    return sorted(str1) == sorted(str2)

# Test the function
string1 = "listen"
string2 = "silent"
if are_anagrams(string1, string2):
    print("The strings are anagrams.")
else:
    print("The strings are not anagrams.")
```

**Q7. Write a Python program to find the index of the first occurrence of a substring in a string**

```python
def find_substring(string, substring):
    return string.find(substring)

string = "hello world"
substring = "world"
index = find_substring(string, substring) # find_substring()
print("Index of the first occurrence:", index)
```

**Q8.Difference between find() and index()**

- find() method is used to find the index of the first occurrence of a substring within a string.If the substring is found, it returns the index of the first occurrence. If not found, it returns -1.
It does not raise an exception if the substring is not found.

```python
st = "hello world"
index = st.find("world")
print("Index of 'world':", index)
```

- The index() method is also used to find the index of the first occurrence of a substring within a string.
Like find(), it returns the index of the first occurrence of the substring.
However, if the substring is not found, it raises a ValueError exception.

```python
string = "hello world"
index = string.index("world")
print("Index of 'world':", index)  #
```
